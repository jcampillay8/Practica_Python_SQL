# Netflix - Employee Retention and Satisfaction

````python
import pandas as pd
import numpy as np
````

````python
df_employee = pd.read_csv('Data/019/employee_satisfaction.csv')
df_employee.head()
````

# Pregunta 1

### El equipo de Recursos Humanos quiere entender el puntaje de satisfacción promedio de los empleados en los diferentes departamentos. ¿Puedes proporcionar el puntaje de satisfacción promedio redondeado hacia abajo al número entero más cercano para cada departamento?

````python
df_department = df_employee.groupby('department_id')['satisfaction_score'].mean().reset_index()

df_department['avg_floor_satisfaction'] = np.floor(df_department['satisfaction_score']).astype(int)

df_department
````

```SQL
SELECT
    department_id,
    FLOOR(avg(satisfaction_score)) AS avg_floor_satisfaction
FROM employee_satisfaction
GROUP BY department_id ;
```

# Pregunta 2

### Además del análisis anterior, el equipo de Recursos Humanos está interesado en conocer el puntaje de satisfacción promedio redondeado hacia arriba al número entero más cercano para cada categoría de trabajo. ¿Puedes calcular esto utilizando los mismos datos?

````python
df_category = df_employee.groupby('job_category_id')['satisfaction_score'].mean().reset_index()

df_category['avg_ceil_satisfaction_category'] = np.ceil(df_category['satisfaction_score']).astype(int)

df_category
````

```SQL
SELECT
    job_category_id,
    CEIL(avg(satisfaction_score)) AS avg_ceil_satisfaction_category
FROM employee_satisfaction
GROUP BY job_category_id;
```

# Pregunta 3 

### El equipo de Recursos Humanos desea un informe consolidado que incluya tanto el promedio de satisfacción redondeado hacia abajo como el redondeado hacia arriba para cada departamento y categoría de trabajo. Por favor, renombra las columnas apropiadamente a 'Floor_Avg_Satisfaction' y 'Ceil_Avg_Satisfaction'.

````python
df_satisfaction = df_employee.groupby(['department_id','job_category_id'])['satisfaction_score'].mean().reset_index()

df_satisfaction['Floor_Avg_Satisfaction'] = np.floor(df_satisfaction['satisfaction_score']).astype(int)
df_satisfaction['Ceil_Avg_Satisfaction'] = np.ceil(df_satisfaction['satisfaction_score']).astype(int)

respuesta3 = df_satisfaction.drop(columns=['satisfaction_score'])

respuesta3
````

```SQL
SELECT
    department_id,
    job_category_id,
    FLOOR(avg(satisfaction_score)) AS Floor_Avg_Satisfaction,
    CEIL(avg(satisfaction_score)) AS Ceil_Avg_Satisfaction
FROM employee_satisfaction
GROUP BY department_id, job_category_id;
```
