# Pregunta 1

### En octubre de 2024, ¿cuál fue el tiempo de instalación más rápido registrado para una actualización de Windows que haya sido exitosa (sin errores)? Esta métrica nos ayudará a establecer un punto de referencia (benchmark) sobre el mejor rendimiento posible en nuestra estrategia de despliegue.

````python
df_octubre = df_update[
    (df_update['installation_date'].between('2024-10-01','2024-10-31')) &
    (df_update['installation_error'] == False)
]

resultado = df_octubre['installation_time_minutes'].min()

print(resultado)
````

```SQL
SELECT
    MIN(installation_time_minutes)
FROM fct_update_installations
WHERE installation_error = FALSE
AND installation_date BETWEEN '2024-10-01' AND '2024-10-31';
```

# Pregunta 2

### En noviembre de 2024, ¿cuántos usuarios únicos instalaron con éxito una actualización de Windows sin encontrar ningún error de instalación? Esta cifra nos ayudará a evaluar la confiabilidad de la actualización.

````python
df_update['installation_date'] = pd.to_datetime(df_update['installation_date'])

nov_success = df_update[
    (df_update['installation_date'].dt.year == 2024) &
    (df_update['installation_date'].dt.month == 11) &
    (df_update['installation_error'] == False)
]

total_usuarios = nov_success['user_id'].nunique()

total_usuarios
````

```SQL
 SELECT
    COUNT(distinct user_id) AS total_unique_users
FROM fct_update_installations
WHERE installation_error = false
AND EXTRACT (YEAR FROM installation_date) = 2024
AND EXTRACT (MONTH FROM installation_date) = 11;
```
