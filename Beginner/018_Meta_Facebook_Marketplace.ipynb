````python
import pandas as pd
import numpy as np
````

````python
df_listing = pd.read_csv('Data/018/listings.csv')
df_listing.head()
````

# Pregunta 1

### ¿Puedes encontrar el precio promedio de los artículos listados en cada categoría en Facebook Marketplace? Queremos entender las tendencias de precios en las diferentes categorías.

````python
df_category_price = df_listing.groupby('category', as_index=False)['price'].mean()

df_category_price
````

```SQL
SELECT
    category,
    avg(price) AS avg_price
FROM Listings
GROUP BY category ;
```

# Pregunta 2

### ¿Qué ciudad tiene el precio promedio más bajo? Esto nos ayudará a identificar las ciudades más asequibles (baratas) para los compradores.

````python
df_city = df_listing.groupby('city')['price'].mean().reset_index()

respuesta2 = df_city.sort_values(by='price', ascending=True).head(1)

respuesta2
````

```SQL
SELECT
    city,
    avg(price) AS avg_price_city
FROM Listings
GROUP BY city
ORDER BY avg_price_city ASC
LIMIT 1;
```
