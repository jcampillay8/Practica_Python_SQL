# Google - Advertising Campaign Performance and ROI Impact

````python
import pandas as pd
import numpy as np
````

````python
df_ads = pd.read_csv('Data/004/ad_campaigns.csv')
df_ads.head()
````

````python
df_ads['start_date'] = pd.to_datetime(df_ads['start_date'])
df_ads_abril = df_ads[
    (df_ads['start_date'].dt.year == 2024) &
    (df_ads['start_date'].dt.month == 4)
]

df_ads_abril.head()
````

````python
resultado = df_ads_abril.sort_values(by='conversions', ascending=False).head(5)['campaign_id']
resultado
````

```SQL
SELECT
    campaign_id
FROM ad_campaigns
WHERE EXTRACT(YEAR FROM start_date) = 2024
AND EXTRACT (MONTH FROM start_date) = 4
order by conversions desc
LIMIT 5;
```

# Pregunta 2

### Identifica los tipos de anuncios (ad_type) únicos que tuvieron más de 100 conversiones y que comenzaron en mayo de 2024.

````python
df_ads['start_date'] = pd.to_datetime(df_ads['start_date'])
mayo_exitoso = df_ads[
    (df_ads['start_date'].dt.year == 2024) &
    (df_ads['start_date'].dt.month == 5) &
    (df_ads['conversions'] > 100)
]

mayo_exitoso.head()
````

````python
tipos_unicos = mayo_exitoso['ad_type'].unique()
tipos_unicos
````

```SQL
SELECT
    distinct ad_type
FROM ad_campaigns
WHERE EXTRACT (YEAR FROM start_date) = 2024
AND EXTRACT (MONTH FROM start_date) = 5
AND conversions > 100;
```

# Pregunta 3

### De las campañas que comenzaron en junio de 2024, encuentra aquellas que no generaron ningún ingreso (revenue). Por favor, muestra los nombres de las campañas (campaign_name) y sus fechas de inicio (start_date).

````python
df_ads['start_date'] = pd.to_datetime(df_ads['start_date'])
df_junio = df_ads[
    (df_ads['start_date'].dt.year == 2024) &
    (df_ads['start_date'].dt.month == 6) &
    (df_ads['revenue'] == 0)
]

df_resultado = df_junio[['campaign_name','start_date']]

df_resultado
````

```SQL
SELECT
    campaign_name,
    start_date
FROM ad_campaigns
WHERE EXTRACT (YEAR FROM start_date) = 2024
AND EXTRACT (MONTH FROM start_date) = 6
AND revenue = 0;
```
